# 🏰 Capstone: Intelligent Document Extractor

**Module 8: Structured Extraction** | **Type: Capstone (Castle)**

---

## Project Overview

Build a production-ready document extraction system that automatically classifies documents, extracts structured data using Pydantic schemas, scores confidence with log probabilities, and processes batches asynchronously.

## Learning Objectives

By the end of this capstone, you will have:

1. **Built** zero-shot and few-shot extraction pipelines for unstructured documents
2. **Implemented** Pydantic schema validation with JSON mode for type-safe extraction
3. **Created** a universal extractor that handles invoices, receipts, resumes, and contracts
4. **Designed** confidence scoring and error recovery using log probabilities
5. **Deployed** async batch processing with progress tracking and reporting
